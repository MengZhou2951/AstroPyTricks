```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(size=10000)
y = np.random.normal(size=10000)
```

```python
plt.scatter(x, y, color='gray')

plt.axhline(0, color='red', linestyle='--')
plt.axvline(0, color='red', linestyle='--')

# if you use hline or vline, you need give xmin/xmax or ymin/ymax
plt.hlines(1, -3, 3, color='blue', linestyle='--')
plt.vlines(1, -3, 3, color='blue', linestyle='--')
```

```python
plt.scatter(x, y, color='gray')

# a line through (-3, -3) and (3, 3)
plt.axline([-3, -3], [3, 3])
```
